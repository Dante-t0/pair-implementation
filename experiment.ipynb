```python
import os
os.environ['OPENAI_API_KEY'] = 'YOUR_OPENAI_API_KEY'
os.environ['TOGETHER_API_KEY'] = "YOUR_TOGETHER_API_KEY"
from pair import PAIR
import pandas as pd
import numpy as np
from openai import OpenAI
```

```python
df = pd.read_csv('harmful_behaviors.csv')
df.head()
```

```python
# index_mapping = [
#     "gpt-3.5-turbo",
#     "mistralai/Mixtral-8x7B-Instruct-v0.1",
#     "Gryphe/MythoMax-L2-13b",
#     "Qwen/Qwen2.5-7B-Instruct-Turbo",
#     "google/gemma-2b-it"
# ]

# columns = ["gpt-3.5-turbo", "Mixtral-8x7B-Instruct-v0.1"]

# results = pd.DataFrame(index=index_mapping, columns=columns)
# results.head()
results = pd.read_csv('results.csv')
results = results.set_index('Unnamed: 0')
results
```

```python
# Experiment 1
sampledPrompts = df['goal'].sample(100)
iterations = 10
successCount = 0
defaultInstance = PAIR()
for objective in sampledPrompts:
    print(f"Running PAIR for objective: {objective}")
    successful_prompt = defaultInstance.pair(iterations, objective)
    
    if successful_prompt:
        successCount += 1
        print("Jailbreak successful!")
    else:
        print("Jailbreak not found.")

print(f"Total successful jailbreaks: {successCount}")
```

```python
results.iloc[0, 0] = successCount / 100
results
```
